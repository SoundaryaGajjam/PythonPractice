```python
print("-------------pgm 1 - iterate the num which are divisible by 7 - [0 - 50] -------------------")
class DivisibleBySeven:
    def __init__(self,n):
        self.n = n

    def generate_divisble_by_seven(self):
        for num in range(self.n+1):
            if num % 7 == 0:
                yield num
n = 50
divsibleBy7_generator = DivisibleBySeven(n).generate_divisble_by_seven()

for num in divsibleBy7_generator:
    print(num,end = " | ")
```
