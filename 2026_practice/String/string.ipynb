```python
print("-------------pgm 1- sord words in alphabetical order-----------------")

str = "Hello, Good evening!!! Today is Saturaday...Have a good day"

words = [word.capitalize() for word in str.split()]

words.sort()
print("Original String :",str)

print("\nSorted array : ",words)
print("\nThe sorted words are : ")
for word in words:
    print(word)
```

```python
print("-------------pgm 2- remove punctuation from a string-----------------")

str = "Hello, Good evening!!! Today is Saturaday...-- and Have a good day"

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
no_punct = ""
print("Original String :",str)

for char in str:
    if char not in punctuations:
        no_punct += char

print("\nAfter removing punctuations : ",no_punct)
```

```python
print("-------------pgm 3- find words which are greater than given lenght --------------")

def find_words(words,k):
    res = []

    for i in words:
        if len(i) > k:
            res.append(i)
    return  res

word_list = ['apple','banana','cherry','date','mango','watermalen','drgon fruit']
k = 7
long_words = find_words(word_list,k)
print(f"Words longer than {k} characters : {long_words}")
```

```python
print("-------------pgm 4- removing ith character from a string --------------")

def remove_char(str,i):
    if i < 0 or i>= len(str):
        print(f"Invalid index {i}. The string remains unchanged.")

    res = str[:i] + str[i+1:]

    return res
str = "Hello world!!!"
i =1

new_str = remove_char(str,i)
print("Original string : ",str)
print(f"String after removing {i}th character: {new_str}")
```

```python
print("-------------pgm 6- check if a given str is binary string or not -------------------")

def is_binary_str(input_str):
    for i in input_str:
        if i not in '01':
            return  False
    return True

input_str = '1001110'


if is_binary_str(input_str):
    print(f"{input_str} is a binary string")
else:
    print(f"{input_str} is not a binary string")
```

```python
print("-------------pgm 7 - find uncommon words from two strings -------------------")

def uncommon_words(str1,str2):
    words1 = set(str1.split())
    words2 = set(str2.split())

    uncommon_set = words1.symmetric_difference(words2)
    uncommon_list = list(uncommon_set)

    return uncommon_list

str1 = "This is the first string"
str2 = "This is the second string"

uncommon = uncommon_words(str1,str2)

print("Uncommon words : ",uncommon)
```

```python
print("-------------pgm 8- find all duplicate chars in strings -------------------")

def find_dups(str):
    char_cnt = {}
    dups = []

    for i in str:
        if i in char_cnt:
            char_cnt[i] += 1
        else:
            char_cnt[i] = 1

    for i, cnt in char_cnt.items():
        if cnt > 1 :
            dups.append(i)

    return  dups
str = "Hello...Good Morning!!!"

dups_char = find_dups(str)
print("Duplicate Characters : ",dups_char)
```

```python
print("-------------pgm 9- chk if a str contains any spl char -------------------")
import  re
def chkSplChar(str):
    print("Input string : ",str)
    pattern = r'[\\\/\.,!-"~!@#$^*()_+={}:;|,?\'"]'

    if re.search(pattern,str):
        return True
    else:
        return False

str = "Hello, Good evening!!! Today is Saturaday...-- and Have a good day"

contains_spl = chkSplChar(str)

if contains_spl :
    print("The str contains special chars")
else:
    print("The str does not contains special chars")
```

```python
print("-------------pgm 12 - count no of letters and digits -------------------")
str = "hello world 123 and 33"

l_cnt = 0
d_cnt = 0

for char in str :
    if char.isalpha():
        l_cnt += 1
    elif char.isdigit():
        d_cnt += 1
print("Input str : ",str)
print("Letter Count : ",l_cnt)
print("Digit Count : ",d_cnt)
```

```python
print("-------------pgm 13 - check the validity of password  -------------------")
import re
def is_valid_password(password):
    if 6 <= len(password) <= 12:
        if re.match(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*([$#@]))",password):
            return True
    return False

password = "ABd1234@1,a F1#,2w3E*,2we3345"
passw = password.split(",")
valid_pass = []

for pwd in passw:
    if is_valid_password(pwd):
        valid_pass.append(pwd)

print("Valid Passwords are : ")
print(",".join(valid_pass))
```
