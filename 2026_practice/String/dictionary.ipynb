```python
import collections
from collections import OrderedDict

print("-------------pgm 1- To extract unique dict values -------------------")

my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 10,
    'd' : 30,
    'e' : 20
}

unq_val = set()
for i in my_dict.values():
    unq_val.add(i)

unq_val_list = list(unq_val)
print("Unique values in the dictonary : ",unq_val_list)
```

```python
print("-------------pgm 2- find the sum of all items in a dict -------------------")

my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 10,
    'd' : 30,
    'e' : 20
}

sum = 0
for i in my_dict.values():
    sum += i

print("Sum of all items in the dict : ",sum)
```

```python
print("-------------pgm 3 - To mergining two dict -------------------")

dict1 = {
    'a' : 10,
    'b' : 20,
}
dict2 = {'c' : 10,
    'd' : 30,
    'e' : 20
}

merge_dict = {**dict1 , **dict2}
print("Merged Dict (using dict unpacking - **dict1 , **dict2) : ",merge_dict)

merge_dict2 = {*dict1 , *dict2}
print("Merged Dict (*dict1 , *dict2) : ",merge_dict2)
```

```python
print("----------------pgm 3 - convert key-values list to flat dict -------------------")

key_val_list = [('a',1),('b',2),('c',3),('d',4)]
flat_dict = {}

for key,value in key_val_list:
    flat_dict[key] = value

print("Flat Dict : ",flat_dict)
```

```python
print("----------------pgm 4 - insertion at the begining in Ordereddict -------------------")
from  collections import OrderedDict

ordered_dict = OrderedDict([('b',2),('c',3),('d',4)])

new_item = ('a',1)
new_ordered_dict = OrderedDict([new_item])

new_ordered_dict.update(ordered_dict)

print("Updated OrderedDict : ",new_ordered_dict)
```

```python
print("----------------pgm 5 - Python dict by key or value -------------------")
my_dict =  {'d': 1, 'c': 2, 'b': 3, 'a': 4}

sorted_dict_by_key = dict(sorted(my_dict.items()))

print("Sorted by keys : ")
for key,val in sorted_dict_by_key.items():
    print(f"{key} : {val}")

sorted_dict_by_key1 = dict(sorted(my_dict.items(),key=lambda  item:item[1]))

print("Sorted by values : ")
print(sorted_dict_by_key1)
```
