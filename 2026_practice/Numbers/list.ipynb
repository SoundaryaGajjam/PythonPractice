```python
print("------------pgm 1- sum of elements in the list------------------")
num = [10,20,30,40]

sum = 0
for i in num:
    sum += i

print("Sum of elements in the list : ",sum)
```

```python
print("------------pgm 2- multiply of elements in the list------------------")
num = [10,20,30,40]

multiply = 1
for i in num:
    multiply *= i

print("Product of elements in the list : ",multiply)
```

```python
print("------------pgm 3- find smallest elements in the list------------------")
num = [10,20,30,5,40]

min = num[0]
for i in num:
   if i < min:
       min = i

print("Smallest of elements in the list : ",min)
```

```python
print("------------pgm 4- find largest elements in the list------------------")
num = [10,20,30,5,40]

max = num[0]
for i in num:
   if i > max:
       max = i

print("Largest of elements in the list : ",max)
```

```python
print("------------pgm 5- find second largest elements in the list------------------")
num = [10,20,30,5,40]

num.sort(reverse=True)

if len(num)>=2:
    sec_large = num[1]
    print("The second largest of elements in the list : ",sec_large)
```

```python
print("------------pgm 6- find N largest elements in the list------------------")
def NthLarge(num,n):
    sort = sorted(num,reverse=True)
    large_ele = sort[:n]
    return large_ele

num = [10,20,30,5,40,3,6,90,100,4,6,39]

n = 3
res = NthLarge(num,n)

print(f"The {n} largest of elements in the list : ",res)
```

```python
print("------------pgm 7- find even & odd nums in the list------------------")
numbers = [10,20,30,5,40,6,7,8,9,1]

even_num = [num for num in numbers if num % 2 == 0]
odd_num = [num for num in numbers if num % 2 != 0]

print("Even nums in the list : ",even_num)
print("Odd nums in the list : ",odd_num)
```

```python
print("------------pgm 8- remove empty list from list------------------")
list_of_list = [[10,20,30],[],[5,40,6],[],[7,8,9,1],[]]

filtered_list = [i for i in list_of_list if i]

print("List after removing empty list : ",filtered_list)
```

```python
print("------------pgm 9- cloning or copying a list------------------")
orginal_list = [1,2,3,4,5]

# mtd1 - using slice operator
cloned_list1 = orginal_list[:]
print("Method 1 - using slice operator - clonned list : ",cloned_list1)

# mtd2 - using list() constructor
#cloned_list2 = list(orginal_list)
#print("Method 2 - using list() constructor - clonned list : ",cloned_list2)

# mtd3 - using list comprehension
cloned_list3 = [item for item in orginal_list]
print("Method 3 - using list comprehension - clonned list : ",cloned_list3)
```

```python
print("------------pgm 10- find occurancesof an element in the list------------------")
def count_occurance(numbers,ele):
    cnt = numbers.count(ele)
    return  cnt

numbers = [1,2,3,4,5,6,2,3,4,12,3,2,34,1,2,2,5,6]
ele_to_cnt = 2
occurances = count_occurance(numbers,ele_to_cnt)

print(f"The element {ele_to_cnt} apperars {occurances} times in the list")
```
