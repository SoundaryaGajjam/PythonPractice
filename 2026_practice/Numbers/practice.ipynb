```python
print("------------pgm 3 - Calculator with 4 basic operation ------------------")

def add(x,y):
    return  x+y

def sub(x,y):
    return x-y

def multiply(x,y):
    return  x*y

def div(x,y):
    return  x/y

print("Select operation :")
print("1. ADD")
print("2. SUB")
print("3. Multiply")
print("4. Divide")

while True:
    choice = input("Enter choice :")
    if choice in ('1','2','3','4'):
        try:
            num1 = 10
            num2 = 20
        except ValueError:
            print(("invalid input : pls enter a no "))
            continue

        if choice == '1':
            print(num1," + ",num2," = ",add(num1,num2))
        elif choice == '2':
            print(num1," - ",num2," = ",sub(num1,num2))
        elif choice == '3':
            print(num1," * ",num2," = ",multiply(num1,num2))
        elif choice == '4':
            print(num1," / ",num2," = ",div(num1,num2))

        nxt_cal = input("let's do nxt calculation y/n")
        if nxt_cal == 'n':
            break
    else:
        print("invalid input")


```

```python
print("------------pgm 4 - Calculate your body mass index ------------------")

def cal_bmi(hgt,wt):
    return  round((wt / hgt ** 2),2)

h = 1.8
w = 70
print("welcome to the BMS calculator : ")
print("height in meter : ",h)
print("weight in kg : ",w)

bmi = cal_bmi(h,w)
print("Your BMI : ",bmi)

if bmi <= 18.5:
    print("your are underweight")
elif 18.5 < bmi <= 24.9:
    print("your weight is normal")
elif 25 < bmi <=29.29:
    print("you are overweight")
else:
    print("you are obese")
```
