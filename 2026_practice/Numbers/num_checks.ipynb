```python
print("-------------------pgm 3 - prime number--------------------------")
num = 27
flag = False

if num == 1:
    print(f"{num} is not a prime num")
elif num > 1:
    for i in range(2,num):
        if (num%i)==0:
            flag=True
            break
    if flag:
        print(f"{num} is not a prime num")
    else:
        print(f"{num} is a prime num")
```

```python
print("-------------------pgm 5 - factorial of a num --------------------------")
num = 4
fact = 1
if num < 0 :
    print("Factorial does not exist for a negative num")
elif num==0:
    print("Factorial of 0 is 1")
else:
    for i in range(1,num+1):
        fact = fact * i
    print(f"The factorial of {num} is {fact}")
```

```python
print("----------------pgm 6 - factorial of a num using recursion --------------------")

def recur_fact(n):
    if n == 1:
        return  n
    else:
        return n*recur_fact(n-1)

num = 5
if num < 0 :
    print("Factorial does not exist for a negative num")
elif num==0:
    print("Factorial of 0 is 1")
else:
    print(f"The factorial of {num} is {recur_fact(num)}")
```

```python
print("-------------------pgm 8 - Fibonacci series --------------------------")
num = 10

n1,n2 = 0,1
cnt = 0
if num <=0 :
    print("Please enter a positive num")
elif num == 1:
    print("fibo seq",n1)
else:
    print("fibo seq")
    while cnt < num:
        print(n1)
        nth = n1+n2
        n1 = n2
        n2 = nth
        cnt = cnt+1
```

```python
print("--------------pgm 9 - Fibonacci series using recursion ---------------------")

def recur_fibo(n):
    if n <=1:
        return n
    else:
        return (recur_fibo(n-1)+recur_fibo(n-2))

num = 7

if num <=0 :
    print("Please enter a positive num")
else:
    print("fibo seq")
    for i in range(num):
        print(recur_fibo(i))
```

```python
print("--------------pgm 10 - chk armstrong num ---------------------")
#153 = 1^3 + 5^3 + 3^3 = 1 + 125 +27 +153

num = 153
num_str = str(num)
print(num_str)
num_digits = len(num_str)
print(num_digits)

sum = 0
tmp = num

while tmp > 0 :
    digit = tmp % 10
    sum += digit ** num_digits
    tmp //= 10

if sum == num :
    print(f"{num} is an armstrong num")
else:
    print(f"{num} is not an armstrong num")
```

```python
print("--------------pgm 11 - chk armstrong num between [1 - 1000]---------------------")
#153 = 1^3 + 5^3 + 3^3 = 1 + 125 +27 +153

lower = 10
upper = 1000

for num in range(lower,upper+1):
    num_digits = len(str(num))

    sum = 0
    tmp = num

    while tmp > 0 :
        digit = tmp % 10
        sum += digit ** num_digits
        tmp //= 10

    if sum == num :
        print(num)
```

```python
print("--------------pgm 13 - LCM ---------------------")

def compute_lcm(x,y):
    if x > y:
        greater = x
    else:
        greater = y
    while(True):
        if((greater % x == 0) and (greater % y == 0)) :
            lcm = greater
            break
        greater += 1
    return  lcm

num1 = 54
num2 = 24

print("The LCM is : ",compute_lcm(num1,num2))
```

```python
print("--------------pgm 14 - HCF ---------------------")

def compute_hcf(x,y):
    if x > y:
        smaller = y
    else:
        smaller = x
    for i in range(1,smaller+1):
        if((x % i == 0) and (y % i == 0)) :
            hcf = i
    return  hcf

num1 = 54
num2 = 24

print("The HCF is : ",compute_hcf(num1,num2))
```
