```python
print("-------------------pgm 3 - prime number--------------------------")
num = 27
flag = False

if num == 1:
    print(f"{num} is not a prime num")
elif num > 1:
    for i in range(2,num):
        if (num%i)==0:
            flag=True
            break
    if flag:
        print(f"{num} is not a prime num")
    else:
        print(f"{num} is a prime num")
```

```python
print("----------------pgm 6 - factorial of a num using recursion --------------------")

def recur_fact(n):
    if n == 1:
        return  n
    else:
        return n*recur_fact(n-1)

num = 5
if num < 0 :
    print("Factorial does not exist for a negative num")
elif num==0:
    print("Factorial of 0 is 1")
else:
    print(f"The factorial of {num} is {recur_fact(num)}")
```

```python
print("--------------pgm 9 - Fibonacci series using recursion ---------------------")

def recur_fibo(n):
    if n <=1:
        return n
    else:
        return (recur_fibo(n-1)+recur_fibo(n-2))

num = 7

if num <=0 :
    print("Please enter a positive num")
else:
    print("fibo seq")
    for i in range(num):
        print(recur_fibo(i))
```

```python
print("--------------pgm 10 - chk armstrong num ---------------------")
#153 = 1^3 + 5^3 + 3^3 = 1 + 125 +27 +153

num = 153
num_str = str(num)
print(num_str)
num_digits = len(num_str)
print(num_digits)

sum = 0
tmp = num

while tmp > 0 :
    digit = tmp % 10
    sum += digit ** num_digits
    tmp //= 10

if sum == num :
    print(f"{num} is an armstrong num")
else:
    print(f"{num} is not an armstrong num")
```

```python
print("--------------pgm 11 - chk armstrong num between [1 - 1000]---------------------")
#153 = 1^3 + 5^3 + 3^3 = 1 + 125 +27 +153

lower = 10
upper = 1000

for num in range(lower,upper+1):
    num_digits = len(str(num))

    sum = 0
    tmp = num

    while tmp > 0 :
        digit = tmp % 10
        sum += digit ** num_digits
        tmp //= 10

    if sum == num :
        print(num)
```

```python
print("--------------pgm 13 - LCM ---------------------")

def compute_lcm(x,y):
    if x > y:
        greater = x
    else:
        greater = y
    while(True):
        if((greater % x == 0) and (greater % y == 0)) :
            lcm = greater
            break
        greater += 1
    return  lcm

num1 = 54
num2 = 24

print("The LCM is : ",compute_lcm(num1,num2))
```

```python
print("--------------pgm 14 - HCF ---------------------")

def compute_hcf(x,y):
    if x > y:
        smaller = y
    else:
        smaller = x
    for i in range(1,smaller+1):
        if((x % i == 0) and (y % i == 0)) :
            hcf = i
    return  hcf

num1 = 54
num2 = 24

print("The HCF is : ",compute_hcf(num1,num2))
```

```python
print("--------------pgm 16 - chk if the given num is a disarium number ------------------")
# 89 = 8^1 + 9^2 = 8+81 = 89

def is_disarium(num):
    num_str = str(num)

    digit_sum = sum(int(i)** (index+1) for index,i in enumerate(num_str))

    return digit_sum == num
try:
    num = 89

    if is_disarium(num):
        print(f"{num} is a Disarium number")
    else:
        print(f"{num} is not a Disarium number")
except ValueError:
    print("Invalid input...Pls enter a valid num")
```

```python
print("--------------pgm 17 - print all disarium num between 1-1000  ------------------")
# 89 = 8^1 + 9^2 = 8+81 = 89

def is_disarium(num):
    num_str = str(num)

    digit_sum = sum(int(i)** (index+1) for index,i in enumerate(num_str))

    return digit_sum == num

disarium_num = [num for num in range(1,1000) if is_disarium(num)]

print("Disarium num between 1-1000 : ")
for num in disarium_num:
    print(num,end=" | ")
```

```python
print("--------------pgm 18 - chk if the given num is a happy number ------------------")
# 19 is happy num
# 1^2 + 9^2 = 1+81 = 82
# 8^2 + 2^2 = 64+4 = 68
# 6^2 + 8^2 = 36+64 = 100
# 1^2 + 0^2 + 0^2 = 1+0+0 = 1
# The process reaches 1, so 19 is happy num

def is_happy(num):
    seen = set() # to store previously seen num

    while num != 1 and num not in seen:
        seen.add(num)
        num = sum(int(i) ** 2 for i in str(num))

    return num == 1
try:
    num = 19

    if is_happy(num):
        print(f"{num} is a Happy number")
    else:
        print(f"{num} is not a Happy number")
except ValueError:
    print("Invalid input...Pls enter a valid num")
```

```python
print("--------------pgm 22 - print all pronic/oblond/rectangular num between 1-100 ------------------")
#P = n * (n+1)
#1 * (1+1) = 2
#2 * (2+1) = 6
#3 * (3+1) = 12
#4 * (4+1) = 20

def is_pronic(num):
    for n in range(1,int(num*0.5)+1):
        if n * (n+1) == num :
            return True
    return False

print("Pronic num between 1 - 100 : ")
for i in range(1,100):
    if is_pronic(i):
        print(i, end = " | ")
```
